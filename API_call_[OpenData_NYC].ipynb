```python
import pandas as pd
import requests

# 1. The API endpoint URL I found on OpenData NYC menu
# Its call "The SODA API" it has 2 version (2 & 3) and thats where I get the
# endpoint (link) for evictions data.
api_url = "https://data.cityofnewyork.us/resource/6z8x-wfk4.json"
#api_url = "https://data.cityofnewyork.us/api/v3/views/6z8x-wfk4/query.json"

# You can add parameters to get more data, as the default is 1000 rows
# For example, to get 50,000 rows:
params = {"$limit": 50000}

# 2. Make the API call
try:
    response = requests.get(api_url, params=params)
    response.raise_for_status() # This will raise an error for bad responses (4xx or 5xx)

    # 3. Convert the JSON response to a Pandas DataFrame
    data = response.json()
    df = pd.DataFrame(data)

    # 4. Initial Exploration - This is where your work begins
    print("DataFrame Info:")
    df.info()

    print("\nFirst 5 Rows:")
    print(df.head())

    print("\nColumn Names:")
    print(df.columns)

except requests.exceptions.RequestException as e:
    print(f"Error making API request: {e}")
```
